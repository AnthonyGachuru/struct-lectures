```python
import numpy as np
import pandas as pd
import os
import sys
import seaborn as sns
sns.set_style('white')
import matplotlib.pyplot as plt
import scipy
```

```python
df = pd.DataFrame()
x = np.linspace(0, 6, 100)
df['x_i'] = x
df['Density'] = scipy.stats.norm.pdf(x, loc=3, scale=1)
fig, ax = plt.subplots(figsize=(6, 4))
sns.lineplot(x='x_i', y='Density', data=df, ax=ax)
ax.set_title('Density of Normal with mu = 3 and sigma=1')
fig.savefig('../figures/normal_density.png', bbox_inches='tight', dpi=200)
```
